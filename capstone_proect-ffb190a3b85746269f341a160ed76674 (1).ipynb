```python
encoder= LabelEncoder()
ohe= OneHotEncoder()

# for training data
df_train['ma']= encoder.fit_transform(df_train['ma'])
dummy_variables = pd.get_dummies(df_train['pc'], prefix='pc', prefix_sep='_')
df_train = pd.concat([df_train, dummy_variables], axis=1)
df_train = df_train.drop('pc', axis=1)

# for testing data
df_test['ma']= encoder.fit_transform(df_test['ma'])
dummy_variables = pd.get_dummies(df_test['pc'], prefix='pc', prefix_sep='_')
df_test = pd.concat([df_test, dummy_variables], axis=1)
df_test = df_test.drop('pc', axis=1)




```

```python
#replacing all nan values with mean values in both train and test data

columns_with_nan = df_train.columns[df_train.isnull().any()]
column_means = df_train.mean()
df_train_filled = df_train.copy()
df_train_filled[columns_with_nan] = df_train_filled[columns_with_nan].fillna(column_means)

columns_with_nan = df_test.columns[df_test.isnull().any()]
column_means = df_test.mean()
df_test_filled = df_test.copy()
df_test_filled[columns_with_nan] = df_test_filled[columns_with_nan].fillna(column_means)
```

```python
from sklearn.preprocessing import MinMaxScaler
X= df_train_filled.drop(columns=['pred'])
y= df_train_filled['pred']
scaler= MinMaxScaler()
X_scaled= scaler.fit_transform(X)
X_scaled_df = pd.DataFrame(X_scaled , columns =X.columns)
X_scaled_df.head()
```

```python
from sklearn.preprocessing import MinMaxScaler
X1= df_test_filled.drop(columns=['index'])
scaler= MinMaxScaler()
X1_scaled= scaler.fit_transform(X1)
X1_scaled_df = pd.DataFrame(X1_scaled , columns =X1.columns)
X1_scaled_df.head()
```

```python
X_train = X_scaled_df
y_train = y
X_test = X1_scaled_df
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
lr = LogisticRegression()
lr.fit(X_train , y_train)
y_pred = lr.predict_proba(X_test)[:,1]
y_preed=(y_pred >= 0.2033).astype(int)
eval_df = df_test_filled.copy()
eval_df['pred'] = y_preed
eval_df.to_csv('test_with_prediction_DTCEf10.csv' , index=False)
```
